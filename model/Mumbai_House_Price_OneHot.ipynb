```python
house_data=house_data.drop(['New/Resale','Lift Available','Car Parking','Maintenance Staff','24x7 Security','Landscaped Gardens','Swimming Pool',"Children's Play Area",'Gas Connection','Jogging Track','Intercom','Clubhouse','Gymnasium','Indoor Games'],axis=1)
```

```python
house_data.shape
```

```python
house_data['Price_per_sqft'] = house_data['Price']/house_data['Area']
```

```python
def removeOutliners(df):
  df_Out = pd.DataFrame()
  for key,subdf in df.groupby('Location'):
    m = np.mean(subdf.Price_per_sqft)
    st = np.std(subdf.Price_per_sqft)
    reduced = subdf[(subdf.Price_per_sqft>(m-st)) & (subdf.Price_per_sqft<=(m+st))]
    df_Out = pd.concat([df_Out,reduced],ignore_index=True)
  return df_Out
house_data = removeOutliners(house_data)
house_data.shape
```

```python
house_data = house_data.drop('Price_per_sqft',axis=1)
```

```python
house_data
```

```python
dummies = pd.get_dummies(house_data.Location)
dummies.head(3)
```

```python
house_data = pd.concat([house_data,dummies],axis='columns')
house_data.head()
```

```python
house_data = house_data.drop('Location',axis='columns')
house_data.head(2)
```

```python
clean_data=house_data.copy()
```

```python
clean_data
```

```python
clean_data.rename(columns={'No. of Bedrooms': 'Bedrooms'}, inplace=True)
```

```python
clean_data
```

```python
X=clean_data.drop(columns=['Price'],axis=1)
Y=clean_data['Price']
```

```python
X
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=10)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(Y.shape,Y_train.shape,Y_test.shape)
```

```python
RF=RandomForestRegressor()
RF.fit(X_train,Y_train)
RF.score(X_test,Y_test)

RF_prediction=RF.predict(X_test)
print('R2 Score :',r2_score(RF_prediction,Y_test)*100)
print('MAE:',mean_absolute_error(RF_prediction,Y_test))
print('MSE:',mean_squared_error(RF_prediction,Y_test))
rmse = sqrt(mean_squared_error(RF_prediction,Y_test))
print('RMSE:',rmse)
```
